**Data Cleaning and Preprocessing**

***1.Handle Missing Values:***

```python
# Check for missing values
print(df.isnull().sum())

# Fill or drop missing values
df = df.fillna(method='ffill')
```

```python
# Check for missing values
print(df1.isnull().sum())
print(df2.isnull().sum())

# Drop or fill missing values if necessary
df1 = df1.dropna()  # or df1.fillna(value)
df2 = df2.dropna()  # or df2.fillna(value)
```

```python
# Rename columns for easier access
df1.rename(columns={'old_name': 'new_name'}, inplace=True)
df2.rename(columns={'old_name': 'new_name'}, inplace=True)
```

```python
# Check the column names of both DataFrames
print("Columns in df1:")
print(df1.columns)
print("\nColumns in df2:")
print(df2.columns)
```

```python
# Remove leading and trailing spaces from column names in both DataFrames
df1.columns = df1.columns.str.strip()
df2.columns = df2.columns.str.strip()

# Verify the column names after stripping spaces
print("Columns in df1 after stripping spaces:")
print(df1.columns)
print("\nColumns in df2 after stripping spaces:")
print(df2.columns)
```

***Merge or Combine Data***

```python
# Merge the two DataFrames on common columns ('Region' and 'Date')
merged_df = pd.merge(df1, df2, on=['Region', 'Date'], suffixes=('_df1', '_df2'))

# Display the merged DataFrame
print("Merged DataFrame:")
print(merged_df.head())
```

```python
import pandas as pd

# Load the two CSV files into DataFrames
df1 = pd.read_csv('Unemployment in India.csv')
df2 = pd.read_csv('Unemployment_Rate_upto_11_2020.csv')

# Display the first few rows of each DataFrame
print("First DataFrame (df1):")
print(df1.head())

print("\nSecond DataFrame (df2):")
print(df2.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the unemployment rate over time from the merged DataFrame
plt.figure(figsize=(10, 6))
sns.lineplot(data=merged_df, x='Date', y='Estimated Unemployment Rate (%)_df1')
plt.title('Unemployment Rate Over Time (DataFrame 1)')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.show()
```

```python
# Example: Create a new column for the difference in unemployment rates between two regions
merged_df['Unemployment Rate Difference'] = merged_df['Estimated Unemployment Rate (%)_df1'] - merged_df['Estimated Unemployment Rate (%)_df2']

# Display the new column
print("Unemployment Rate Difference:")
print(merged_df[['Region', 'Date', 'Unemployment Rate Difference']].head())
```

```python
# Pivot the data to create a matrix for heatmap
heatmap_data = merged_df.pivot_table(values='Estimated Unemployment Rate (%)_df1', index='Region', columns='Date')

plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap="YlGnBu", annot=False)
plt.title('Heatmap of Unemployment Rate by Region')
plt.show()
```
